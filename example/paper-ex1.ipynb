```python
import matplotlib . pyplot as plt
def fib ( n ) :
    return n if n < 2 else fib (n -1) + fib (n -2)
def fact ( n ) :
    return n * fact (n -1) if n > 1 else 1
x = 35
```

```python
fib_x = fib(x)
```

```python
fact_x = fact(x)
```

```python
plt.bar(["fib(x)", "fact(x)"], [fib_x, fact_x])
for i, val in enumerate([fib_x, fact_x]):
    plt.text(i, val*1.02, str(val), ha='center')
plt.savefig("./data/plot.png")
```
