```python
import numpy as np
import time
import os
output_file_dir = './data'
```

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

start_time = time.time()
for i in range(3):
    random_numbers = np.random.randint(1, 10000, 10000)
    sorted_numbers = quick_sort(random_numbers)
end_time = time.time()

res1 = (end_time - start_time) / 3
```

```python

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

start_time = time.time()
for i in range(3):  # Repeat the sort 3 times to find the average time
    random_numbers = np.random.randint(1, 10000, 10000)
    merge_sort(random_numbers)
end_time = time.time()

res2 = (end_time - start_time) / 3
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >=0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

start_time = time.time()
for i in range(3):
    random_numbers = np.random.randint(1, 10000, 10000)
    insertion_sort(random_numbers)
end_time = time.time()

res3 = (end_time - start_time) / 3

```

```python
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[min_idx] > arr[j]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

start_time = time.time()
for i in range(3):  # Repeat the sort 3 times to find the average time
    random_numbers = np.random.randint(1, 10000, 10000)
    selection_sort(random_numbers)
end_time = time.time()

res4 = (end_time - start_time) / 3
```

```python
#R
if (!require(jsonlite, quietly = TRUE)) {
    install.packages("jsonlite", repos = "http://cran.r-project.org")
    library(jsonlite)
}

# Bubble Sort function definition
bubble_sort <- function(arr) {
    n <- length(arr)
    for (i in 1:(n-1)) {
        for (j in 1:(n-i)) {
            if (arr[j] > arr[j+1]) {
                temp <- arr[j]
                arr[j] <- arr[j+1]
                arr[j+1] <- temp
            }
        }
    }
    return(arr)
}

# Timing the execution of the Bubble Sort
start_time <- Sys.time()

# Perform Bubble Sort 3 times on arrays of 10,000 random numbers
sorted_results <- list()
for (i in 1:3) {
    random_numbers <- sample(1:10000, 10000, replace = TRUE)
    sorted_numbers <- bubble_sort(random_numbers)
    sorted_results[[i]] <- sorted_numbers
}

end_time <- Sys.time()

# Calculate the average function run time
average_run_time <- difftime(end_time, start_time, units = "secs") / 3

# Prepare directory for data storage
data_dir <- "./data"
if (!dir.exists(data_dir)) {
    dir.create(data_dir)
}

# Prepare data for JSON file
results <- list(average_run_time = as.numeric(average_run_time))
json_data <- toJSON(results, pretty = TRUE, auto_unbox = TRUE)

# Write to JSON file
file_path <- file.path(data_dir, "res5_from_r.json")
writeLines(json_data, file_path)
```
