```python
import numpy as np
import time
```

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

for i in range(3):
    random_numbers = np.random.randint(1, 10000, 10000)
    sorted_numbers = quick_sort(random_numbers)

current_time = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime())

with open("quick_sort_times.txt", "w") as file:
    file.write(current_time)

```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

for i in range(3):
    random_numbers = np.random.randint(1, 10000, 10000)
    bubble_sort(random_numbers)
current_time = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime())



with open("bubble_sort_times.txt", "w") as file:
    file.write(current_time)

```

```python

def fibonacci(n):
    if n <= 1:
        return n
    else:
        return(fibonacci(n-1) + fibonacci(n-2))

times = []
for i in range(3):  
    start_time = time.time()
    fibonacci_numbers = [fibonacci(i) for i in range(40)]
    times.append(time.time() - start_time)

average_time = sum(times) / len(times)
with open("fibonacci_times.txt", "w") as file:
    file.write(f"Average execution time: {average_time} seconds")
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >=0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

times = []
for i in range(3):
    random_numbers = np.random.randint(1, 10000, 10000)
    start_time = time.time()
    insertion_sort(random_numbers)
    times.append(time.time() - start_time)

average_time = sum(times) / len(times)
with open("insertion_sort_times.txt", "w") as file:
    file.write(f"Average execution time: {average_time} seconds")

```

```python
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[min_idx] > arr[j]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

times = []
for i in range(3):  # Repeat the sort 3 times to find the average time
    random_numbers6 = np.random.randint(1, 10000, 10000)
    start_time = time.time()
    selection_sort(random_numbers6)
    times.append(time.time() - start_time)

average_time = sum(times) / len(times)
with open("selection_sort_times.txt", "w") as file:
    file.write(f"Average execution time: {average_time} seconds")
```

```python


def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

times = []
for i in range(3):  # Repeat the sort 3 times to find the average time
    random_numbers7 = np.random.randint(1, 10000, 10000)
    start_time = time.time()
    merge_sort(random_numbers7)
    times.append(time.time() - start_time)

average_time = sum(times) / len(times)
with open("merge_sort_times.txt", "w") as file:
    file.write(f"Average execution time: {average_time} seconds")
```

```python

def heapify(arr, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and arr[l] > arr[largest]:
        largest = l

    if r < n and arr[r] > arr[largest]:
        largest = r

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)

    # Build a maxheap.
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    # One by one extract elements
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]  # swap
        heapify(arr, i, 0)

times = []
for i in range(3):  # Repeat the sort 3 times to find the average time
    random_numbers8 = np.random.randint(1, 10000, 10000)
    start_time = time.time()
    heap_sort(random_numbers8)
    times.append(time.time() - start_time)

average_time = sum(times) / len(times)
with open("heap_sort_times.txt", "w") as file:
    file.write(f"Average execution time: {average_time} seconds")
```
