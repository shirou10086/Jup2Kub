## Part 1: Data Preprocessing

```python
df = pd.read_csv('./data/musicData.csv')
df.head()
```

```python
df.describe() # note that this only describes purely numerical data
```

```python
# note that there are some columns that should be numerical but didn't show up in the df.discribe()
df.info()
```

For cleaning the numerical data, we need to
- the Dtype of tempo column should not be 'object', meaning there are missing values, so we use the avarage to replace
- according to the desribe print, the 'min' entry for duration is -1, which are missing values, so we use average to replace
- we need to drop NA, while there are 50005 lines but only 50000 data
- the IDs and the obtained data by Spotify doesn't make any logical sense for classification, so I will delete them. The artist/song names are also required to be deleted according to the specification.

```python
# remove NA entries
df.dropna(inplace=True)
# remove the columns needed to be excluded
df.drop(['instance_id', 'artist_name', 'track_name', 'obtained_date'], axis=1, inplace=True)
# inputation for tempo and duration
df['tempo'] = pd.to_numeric(df['tempo'], errors='coerce')
meanTempo = df['tempo'].mean()
df['tempo'].fillna(meanTempo, inplace=True)
meanDuration = df.loc[df['duration_ms'] > 0, 'duration_ms'].mean()
df.loc[df['duration_ms'] <= 0, 'duration_ms'] = meanDuration
```

For encoding the catagorical data, we are required to
- dummy code 'mode' to all zeros or ones
- as I also play piano, **based on my knowledge on music theory, the adjacent keys like C and D have completely different emotional expressions, therefore, I don't think it's a good idea to transform the keys into numerical values as the locality of keys doesn't imply locality of emotions**, rather, I will code them to be multiple columns of 1/0 vlaues

```python
# dummy code 'mode'
dummyMode = pd.get_dummies(df['mode'])
df['mode'] = dummyMode.iloc[:, 0]
# transform keys into dummy columns
dummyKey = pd.get_dummies(df['key'])
df.drop('key', axis=1, inplace=True)
df = pd.concat([df.iloc[:, :-1], dummyKey, df.iloc[:, -1]], axis=1)
df.head()
```

Then, we need to normalize the numeric values!

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
#df['popularity', 'acousticness', 'danceability', 'duration_ms', 'energy', 'instrumentalness', 'liveness', 'loudness']
numericCols = ['popularity', 'acousticness', 'danceability', 'duration_ms', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence']
df[numericCols] = StandardScaler().fit_transform(df[numericCols])
```

```python
df.columns
```

Finish preprocessing

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# also encode labels to numeric values
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
y_strings = encoder.classes_
```

1. PCA

```python
X_tsne = X_tsne_original
# use distance to filter out outliers
centroid = X_tsne.mean(axis=0)
distances = cdist(X_tsne, [centroid]).ravel()
plot_indices = distances <= 50
X_tsne = X_tsne[plot_indices]
y_tsne = y_encoded[plot_indices]

# plot the reduced X in 3 dimention space
fig = plt.figure(figsize=(21, 14))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(X_tsne[:, 0], X_tsne[:, 1], X_tsne[:, 2], c=y_tsne, s=1, alpha = 0.6)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

plt.show()
```

## Part3: Classification
**But do train test split first**, note that the 'stratify' parameter specified the 4500 to 500 ratio for each class

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.1, stratify=y, random_state=12598673)

# also do the imports here
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.metrics import roc_curve, auc
```

### Attempt 1: SVM

```python
# train SVM
svm = SVC()
svm.fit(X_train, y_train)
prediction = svm.predict(X_test)

accuracy = accuracy_score(y_test, prediction)
print("Accuracy:", accuracy)
```

### Attempt 2: AdaBoost

```python
# train adaBoost
ab = AdaBoostClassifier(random_state=12598673)
ab.fit(X_train, y_train)
prediction = ab.predict(X_test)

accuracy = accuracy_score(y_test, prediction)
print("Accuracy:", accuracy)
```

#### Attempt 3: 3-layer Neural Network

```python
def init_model(activation, D, H, C):
    if activation == 0:
        af = nn.Identity()
    elif activation == 1:
        af = nn.ReLU()
    elif activation == 2:
        af = nn.Sigmoid()
    elif activation == 3:
        af = nn.Tanh()
    elif activation == 4:
        af = nn.Softplus()

    model = nn.Sequential(
        nn.Linear(D, H),
        af,
        nn.Linear(H, H),
        af,
        nn.Linear(H, H),
        af,
        nn.Linear(H, C),
    )
    return model
```

```python
def train_and_test(activation, D, H, C, X_train, X_test, y_train, y_test, learning_rate):
    model = init_model(activation, D, H, C)
    criterion = torch.nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate, weight_decay=lambda_l2)

    accuracyHistory = []

    for t in range(4000):
        prediction = model(X_train)
        loss = criterion(prediction, y_train)

        score, predicted = torch.max(prediction, 1)
        acc = (y_train == predicted).sum().float() / len(y_train)
    
        optimizer.zero_grad()

        loss.backward()
        optimizer.step()

        predicted_labels = torch.argmax(prediction, dim=1)
        accuracy = (predicted_labels == y_train).sum().item() / len(y_train)
        accuracyHistory.append(accuracy)
    
    prediction = model(X_test)
    prediction = F.softmax(model(X_test), dim=1)
    return(roc_auc_score(y_test, prediction.cpu().detach(), multi_class='ovr', average=None)), accuracyHistory, model
```

```python
# setup hyper parameters
torch.manual_seed(12598673)
D = X_train.shape[1] # dimentions, number of features
C = 10 # num_classes
H = 100 # num_hidden_units
learning_rate = 0.2
lambda_l2 = 1e-6
X_train_tensor = torch.tensor(X_train.values, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)
score, history, trained_model = train_and_test(1, D, H, C, X_train_tensor, X_test_tensor, y_train_tensor, y_test, learning_rate)
print(score)
```

```python
# draw plot for the accuracy history in each iteration
iters = np.linspace(1, 4000, 4000)
plt.plot(iters, history)
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.show()
```
